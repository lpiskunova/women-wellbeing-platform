```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(
    style="whitegrid",
    font_scale=1.0,
    rc={
        "axes.spines.top": False,
        "axes.spines.right": False
    }
)

purple_palette = [
    "#DEC2FF",
    "#9671C9",
    "#7C3AED",
    "#A855F7",
    "#EC4899",
    "#752AAD",
    "#7518F5",
    "#7B26FF",
]

single_purple = "#7C3AED"

from matplotlib.colors import LinearSegmentedColormap
purple_cmap = LinearSegmentedColormap.from_list(
    "purple_cmap",
    ["#FDF2FF", "#C4B5FD", "#7C3AED", "#4C1D95"]
)
```

```python
df = pd.read_csv("unodc-femicide.csv")
df.head(15)
```

```python
df = df.dropna()
df = df.reset_index(drop=True)

df.info()
df.isna().sum()
```

```python
df['Year'] = df['Year'].astype(int)
df.info()
```

```python
df = df.rename(columns={
    'Iso3_code': 'iso3',
    'Country': 'country',
    'Region': 'region',
    'Subregion': 'subregion',
    'Indicator': 'indicator',
    'Dimension': 'dimension',
    'Category': 'category',
    'Sex': 'sex',
    'Age': 'age',
    'Year': 'year',
    'Unit of measurement': 'unit',
    'VALUE': 'value',
    'Source': 'source'
})
```

```python
df.head()
df.info()
```

```python
victims = df[df['indicator'] == 'Victims of intentional homicide'].copy()
suspects = df[df['indicator'] == 'Persons arrested/suspected for intentional homicide'].copy()

victims['indicator'].unique(), suspects['indicator'].unique()
```

```python
df['unit'].value_counts()
```

```python
victims_counts = victims[victims['unit'] == 'Counts'].copy()
victims_rates  = victims[victims['unit'] == 'Rate per 100,000 population'].copy()

print(victims_counts['unit'].value_counts())
print(victims_rates['unit'].value_counts())
```

```python
# Total female victims per year (all countries, all ages)
global_counts_by_year = (
    victims_counts
    .groupby('year', as_index=False)['value']
    .sum()
    .rename(columns={'value': 'total_victims'})
)

display(global_counts_by_year.head())

plt.figure(figsize=(10, 5))
plt.plot(global_counts_by_year['year'],
         global_counts_by_year['total_victims'],
         marker='o', color=single_purple,)
plt.title('Total number of female homicide victims by year')
plt.xlabel('Year')
plt.ylabel('Number of victims')
plt.grid(True)
plt.show()
```

```python
region_counts_by_year = (
    victims_counts
    .groupby(['region', 'year'], as_index=False)['value']
    .sum()
    .rename(columns={'value': 'total_victims'})
)

display(region_counts_by_year.head())

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=region_counts_by_year,
    x='year',
    y='total_victims',
    hue='region',
    marker='o',
    palette=purple_palette
)
plt.title('Female homicide victims by region and year')
plt.xlabel('Year')
plt.ylabel('Number of victims')
plt.legend(title='Region')
plt.grid(True)
plt.show()
```

```python
year_focus = victims_counts['year'].max()
year_focus
```

```python
age_dist = (
    victims_counts[victims['year'] == year_focus]
    .groupby('age', as_index=False)['value']
    .sum()
    .rename(columns={'value': 'total_victims'})
)

# Order age groups
age_order = ['0-9', '10 -14', '15 -17', '18-19', '20-24',
             '25-29', '30-44', '45-59', '60 and older']

age_dist['age'] = pd.Categorical(age_dist['age'],
                                 categories=age_order,
                                 ordered=True)
age_dist = age_dist.sort_values('age')

display(age_dist)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=age_dist, x='age', y='total_victims', palette=purple_palette)
plt.title(f'Distribution of female homicide victims by age group, {year_focus}')
plt.xlabel('Age group')
plt.ylabel('Number of victims')
plt.xticks(rotation=45)
plt.show()
```

```python
country_counts_year = (
    victims_counts[victims['year'] == year_focus]
    .groupby(['iso3', 'country'], as_index=False)['value']
    .sum()
    .rename(columns={'value': 'total_victims'})
)

# Top 10 countries
top10_countries = country_counts_year.sort_values(
    'total_victims', ascending=False
).head(10)

display(top10_countries)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top10_countries,
    x='total_victims',
    y='country',
    palette=purple_palette
)
plt.title(f'Top 10 countries by number of female homicide victims, {year_focus}')
plt.xlabel('Number of victims')
plt.ylabel('Country')
plt.show()
```

```python
coverage_by_year = (
    victims_counts
    .groupby('year', as_index=False)
    .agg(
        n_records=('value', 'size'),
        n_countries=('country', 'nunique')
    )
)

coverage_by_year
```

```python
plt.figure(figsize=(10, 5))
plt.plot(coverage_by_year['year'],
         coverage_by_year['n_countries'],
         marker='o', color=single_purple)
plt.title('Number of reporting countries by year')
plt.xlabel('Year')
plt.ylabel('Number of countries')
plt.grid(True)
plt.show()
```

```python
country_coverage = (
    victims_counts
    .groupby(['iso3', 'country'], as_index=False)
    .agg(
        n_years=('year', 'nunique'),
        first_year=('year', 'min'),
        last_year=('year', 'max'),
        total_victims=('value', 'sum')
    )
)

country_coverage.head()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(country_coverage['n_years'], bins=10, edgecolor='black', color=single_purple)
plt.title('Distribution of number of reporting years per country')
plt.xlabel('Number of years with data')
plt.ylabel('Number of countries')
plt.grid(True)
plt.show()
```

```python
# total victims by year & region
region_totals = (
    victims_counts
    .groupby(['year', 'region'], as_index=False)['value']
    .sum()
    .rename(columns={'value': 'region_victims'})
)

# global total by year
global_totals = (
    victims_counts
    .groupby('year', as_index=False)['value']
    .sum()
    .rename(columns={'value': 'global_victims'})
)

region_share = region_totals.merge(global_totals, on='year', how='left')
region_share['share_pct'] = region_share['region_victims'] / region_share['global_victims'] * 100

region_share.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=region_share,
    x='year',
    y='share_pct',
    hue='region',
    marker='o',
    palette=purple_palette
)
plt.title('Regional share of female homicide victims (%)')
plt.xlabel('Year')
plt.ylabel('Share of victims, %')
plt.legend(title='Region')
plt.grid(True)
plt.show()
```

```python
early_period = victims_counts[victims_counts['year'].between(2005, 2009)]
late_period  = victims_counts[victims_counts['year'].between(2019, 2023)]

early_by_country = (
    early_period
    .groupby('country', as_index=False)['value']
    .sum()
    .rename(columns={'value': 'early_victims'})
)

late_by_country = (
    late_period
    .groupby('country', as_index=False)['value']
    .sum()
    .rename(columns={'value': 'late_victims'})
)

change = early_by_country.merge(late_by_country, on='country', how='inner')
change = change[change['early_victims'] > 0]
change['diff'] = change['late_victims'] - change['early_victims']
change['pct_change'] = change['diff'] / change['early_victims'] * 100

# countries with the highest growth
top_increase = change.sort_values('diff', ascending=False).head(10)

# countries with the largest decline
top_decrease = change.sort_values('diff', ascending=True).head(10)

top_increase, top_decrease
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_increase,
    x='diff',
    y='country',
    palette=purple_palette
)
plt.title('Top 10 countries by increase in female homicide victims\n(20192023 vs 20052009)')
plt.xlabel('Change in number of victims')
plt.ylabel('Country')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_decrease,
    x='diff',
    y='country',
    palette=purple_palette
)
plt.title('Top 10 countries by decrease in female homicide victims\n(20192023 vs 20052009)')
plt.xlabel('Change in number of victims')
plt.ylabel('Country')
plt.show()
```

```python
year_focus = victims_counts['year'].max()

top_countries_list = top10_countries['country'].tolist()

top_countries_age = (
    victims_counts[
        (victims_counts['year'] == year_focus) &
        (victims_counts['country'].isin(top_countries_list))
    ]
    .groupby(['country', 'age'], as_index=False)['value']
    .sum()
    .rename(columns={'value': 'total_victims'})
)

age_order = ['0-9', '10-14', '15-17', '18-19',
             '20-24', '25-29', '30-44', '45-59', '60 and older']

top_countries_age['age'] = pd.Categorical(top_countries_age['age'],
                                          categories=age_order,
                                          ordered=True)
top_countries_age = top_countries_age.sort_values(['country', 'age'])
```

```python
g = sns.catplot(
    data=top_countries_age,
    x='age',
    y='total_victims',
    col='country',
    col_wrap=3,
    kind='bar',
    height=3.5,
    sharey=False,
    palette=purple_palette
)
g.set_titles('{col_name}')
g.set_axis_labels('Age group', 'Number of victims')
for ax in g.axes.flatten():
    ax.tick_params(axis='x', rotation=45)
plt.suptitle(f'Age distribution of female homicide victims in top 10 countries, {year_focus}', y=1.03)
plt.show()
```

```python
year_focus = victims_counts['year'].max()

subregion_year = (
    victims_counts[victims_counts['year'] == year_focus]
    .groupby(['region', 'subregion'], as_index=False)['value']
    .sum()
    .rename(columns={'value': 'total_victims'})
)

subregion_year
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=subregion_year,
    x='total_victims',
    y='subregion',
    hue='region',
    palette=purple_palette
)
plt.title(f'Female homicide victims by subregion, {year_focus}')
plt.xlabel('Number of victims')
plt.ylabel('Subregion')
plt.legend(title='Region')
plt.show()
```

```python
# choose year for heatmap
year_focus = 2020

age_order = ['0-9', '10-14', '15-17', '18-19',
             '20-24', '25-29', '30-44', '45-59', '60 and older']

# total victims by region & age for the selected year
heat_data = (
    victims_counts[victims_counts['year'] == year_focus]
    .groupby(['region', 'age'], as_index=False)['value']
    .sum()
    .rename(columns={'value': 'total_victims'})
)

# share of each age group within region (%)
heat_data['region_total'] = heat_data.groupby('region')['total_victims'].transform('sum')
heat_data['share_pct'] = heat_data['total_victims'] / heat_data['region_total'] * 100

# make age an ordered categorical to control column order
heat_data['age'] = pd.Categorical(heat_data['age'],
                                  categories=age_order,
                                  ordered=True)

# pivot for heatmap
heat_pivot = heat_data.pivot_table(
    index='region',
    columns='age',
    values='share_pct'
)

plt.figure(figsize=(10, 6))
sns.heatmap(
    heat_pivot,
    annot=True, fmt='.1f',
    cmap=purple_cmap
)
plt.title(f'Age structure of female homicide victims by region, {year_focus}\n(share within region, %)')
plt.xlabel('Age group')
plt.ylabel('Region')
plt.show()
```

```python
year_a = 2010
year_b = 2020

vt_a = (
    victims_counts[victims_counts['year'] == year_a]
    .groupby(['country', 'region'], as_index=False)['value']
    .sum()
    .rename(columns={'value': 'victims_a'})
)

vt_b = (
    victims_counts[victims_counts['year'] == year_b]
    .groupby(['country', 'region'], as_index=False)['value']
    .sum()
    .rename(columns={'value': 'victims_b'})
)

scatter_df = vt_a.merge(
    vt_b[['country', 'region', 'victims_b']],
    on=['country', 'region'],
    how='inner'
)

scatter_df.head()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(
    data=scatter_df,
    x='victims_a',
    y='victims_b',
    hue='region',
    palette=purple_palette
)

max_val = max(scatter_df['victims_a'].max(), scatter_df['victims_b'].max())
plt.plot([0, max_val], [0, max_val], 'k--', linewidth=1)

plt.title(f'Female homicide victims: {year_b} vs {year_a}')
plt.xlabel(f'Number of victims, {year_a}')
plt.ylabel(f'Number of victims, {year_b}')
plt.grid(True)
plt.legend(title='Region')
plt.show()
```

```python
# Average rate per 100,000 population by year (all ages, all countries)
rate_global_by_year = (
    victims_rates
    .groupby('year', as_index=False)['value']
    .mean()
    .rename(columns={'value': 'mean_rate'})
)

display(rate_global_by_year.head())

plt.figure(figsize=(10, 5))
plt.plot(rate_global_by_year['year'],
         rate_global_by_year['mean_rate'],
         marker='o', color=single_purple)
plt.title('Average female homicide rate per 100,000 population by year')
plt.xlabel('Year')
plt.ylabel('Rate per 100,000 population')
plt.grid(True)
plt.show()
```

```python
region_rates_by_year = (
    victims_rates
    .groupby(['region', 'year'], as_index=False)['value']
    .mean()
    .rename(columns={'value': 'rate_per_100k'})
)

display(region_rates_by_year.head())

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=region_rates_by_year,
    x='year',
    y='rate_per_100k',
    hue='region',
    marker='o',
    palette=purple_palette
)
plt.title('Female homicide rate per 100,000 population by region and year')
plt.xlabel('Year')
plt.ylabel('Rate per 100,000 population')
plt.legend(title='Region')
plt.grid(True)
plt.show()
```

```python
year_focus = victims_rates['year'].max()

country_rate_year = (
    victims_rates[victims_rates['year'] == year_focus]
    .groupby(['iso3', 'country'], as_index=False)['value']
    .mean()
    .rename(columns={'value': 'rate_per_100k'})
)

top10_rate_countries = (
    country_rate_year
    .sort_values('rate_per_100k', ascending=False)
    .head(10)
)

display(top10_rate_countries)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=top10_rate_countries,
    x='rate_per_100k',
    y='country',
    palette=purple_palette
)
plt.title(f'Top 10 countries by female homicide rate, {year_focus}')
plt.xlabel('Rate per 100,000 population')
plt.ylabel('Country')
plt.show()
```
